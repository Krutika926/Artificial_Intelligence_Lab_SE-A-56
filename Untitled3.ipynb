## Implement BFS

### Subtask:
Write a Python function to implement the Breadth-First Search (BFS) algorithm. The function should take a graph and a starting node as input and return the order of visited nodes.


**Reasoning**:
I will define the BFS function as per the instructions provided in the subtask, including initialization of visited nodes, queue, and visited set, and then implement the BFS logic using a while loop.



````python
from collections import deque

def bfs(graph, start_node):
    visited_nodes = []  # To store the traversal order
    queue = deque([start_node])  # Initialize a queue and add the start node
    visited = {start_node}  # To keep track of visited nodes

    while queue:
        current_node = queue.popleft()  # Dequeue a node from the front
        visited_nodes.append(current_node)  # Add to traversal order

        # Iterate through neighbors
        for neighbor in graph[current_node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return visited_nodes

print("BFS function defined successfully.")
````

## Implement DFS

### Subtask:
Write a Python function to implement the Depth-First Search (DFS) algorithm. The function should take a graph and a starting node as input and return the order of visited nodes.

**Reasoning**:
Now that the DFS subtask has been introduced, I will define the Python function for Depth-First Search. This function will use a stack (implicit with recursion or explicit) and a set for visited nodes to track the traversal order, similar to the BFS implementation but with a different traversal mechanism.



````python
def dfs(graph, start_node):
    visited_nodes = []  # To store the traversal order
    stack = [start_node]  # Initialize a stack and add the start node
    visited = {start_node}  # To keep track of visited nodes

    while stack:
        current_node = stack.pop()  # Pop a node from the top of the stack
        visited_nodes.append(current_node)  # Add to traversal order

        # Iterate through neighbors (in reverse to simulate consistent left-to-right exploration for non-recursive DFS)
        # Or, just iterate and push, the order of exploration depends on the stack. Reversing for consistency.
        for neighbor in sorted(graph[current_node], reverse=True):
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)
    return visited_nodes

print("DFS function defined successfully.")
````

### Summary of BFS and DFS Differences

Based on the implementations and execution results, here's a summary of the key differences and characteristics between Breadth-First Search (BFS) and Depth-First Search (DFS):

**1. Traversal Order:**
*   **BFS:** Explores all neighbors at the current depth level before moving on to nodes at the next depth level. It explores the graph layer by layer.
    *   *Observed Traversal (`bfs_traversal`):* `['A', 'B', 'C', 'D', 'E', 'F', 'G']` - Shows a systematic exploration of nodes closer to the start node first.
*   **DFS:** Explores as far as possible along each branch before backtracking. It goes deep into one path before exploring other paths.
    *   *Observed Traversal (`dfs_traversal`):* `['A', 'B', 'D', 'E', 'F', 'G', 'C']` - Shows it went deep down the 'B' path (A -> B -> D -> E -> F -> G) before coming back to 'C'.

**2. Data Structure:**
*   **BFS:** Uses a **queue** (First-In, First-Out or FIFO) to manage the nodes to visit. This ensures that nodes at the same level are visited before moving to the next level.
*   **DFS:** Uses a **stack** (Last-In, First-Out or LIFO), either explicitly or implicitly through recursion, to manage the nodes to visit. This drives the deep exploration characteristic.

**3. Space Complexity:**
*   **BFS:** In the worst case, BFS might store all nodes at the widest level of the graph in the queue. For a dense graph, this can be significant.
*   **DFS:** In the worst case, DFS might store all nodes along the longest path in the stack. For a deep, narrow graph, this can be significant.

**4. Applications:**
*   **BFS:** Often used for finding the shortest path in an unweighted graph, finding the shortest distance between two nodes, or for network broadcast.
*   **DFS:** Often used for cycle detection, topological sorting, finding connected components, or solving puzzles with a single solution (e.g., mazes).

## Implement A* Algorithm

### Subtask:
Write a Python function for the A* search algorithm. This function should take the grid, start, and end points as input, use the heuristic function, and return the shortest path found. It will involve managing an open set (priority queue), a closed set, and calculating 'g' and 'f' scores.


**Reasoning**:
I will now implement the A* search algorithm function following the detailed instructions provided, including importing `heapq`, initializing necessary data structures, and implementing the main search loop with path reconstruction.



````python
import heapq

def a_star_search(grid, start, end, heuristic_func):
    rows, cols = grid.shape

    # Helper function to check if a node is valid and not an obstacle
    def is_valid(r, c):
        return 0 <= r < rows and 0 <= c < cols and grid[r, c] == 0

    # Reconstruct path from came_from dictionary
    def reconstruct_path(came_from, current):
        path = []
        while current in came_from:
            path.append(current)
            current = came_from[current]
        path.append(start) # Add the start node
        return path[::-1] # Reverse to get path from start to end

    # 1. Initialize open_set (priority queue): (f_score, node)
    #   The f_score for the start node will be its heuristic value since its g_score is 0.
    open_set = [(heuristic_func(start, end), start)]

    # 2. Initialize came_from to reconstruct the path
    came_from = {}

    # 3. Initialize g_score (cost from start to current node)
    g_score = { (r, c): float('inf') for r in range(rows) for c in range(cols) }
    g_score[start] = 0

    # 4. Initialize f_score (g_score + heuristic_score)
    f_score = { (r, c): float('inf') for r in range(rows) for c in range(cols) }
    f_score[start] = heuristic_func(start, end)

    # Possible movements (up, down, left, right)
    movements = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    while open_set:
        # Pop the node with the lowest f-score
        current_f, current_node = heapq.heappop(open_set)

        # If we reached the end node, reconstruct and return the path
        if current_node == end:
            return reconstruct_path(came_from, current_node)

        # For each neighbor of the current node
        for dr, dc in movements:
            neighbor_r, neighbor_c = current_node[0] + dr, current_node[1] + dc
            neighbor = (neighbor_r, neighbor_c)

            # Check if neighbor is valid and not an obstacle
            if is_valid(neighbor_r, neighbor_c):
                # Calculate tentative g_score
                tentative_g_score = g_score[current_node] + 1  # Cost to move is 1

                # If this path to neighbor is better than any previous one
                if tentative_g_score < g_score[neighbor]:
                    came_from[neighbor] = current_node
                    g_score[neighbor] = tentative_g_score
                    f_score[neighbor] = tentative_g_score + heuristic_func(neighbor, end)

                    # If neighbor is not already in open_set, add it
                    # heapq does not have an efficient way to check membership or update priority
                    # so we just add duplicates; the one with lower f-score will be popped first.
                    # We rely on the check `if current_node == end` to stop processing old entries.
                    # A more robust solution might use a dictionary for open_set lookup too.
                    heapq.heappush(open_set, (f_score[neighbor], neighbor))

    return None # No path found

print("A* search function defined successfully.")
````

*   By prioritizing nodes with the lowest `f_score`, A* ensures that it explores paths that are both *short so far* and *likely to lead quickly to the goal*. ## Define Game State and Rules

### Subtask:
Define a simple game board representation (e.g., Tic-Tac-Toe) and implement functions to check for wins, draws, and available moves.


**Reasoning**:
I will define the Tic-Tac-Toe game board and implement the required `check_win`, `check_draw`, and `get_available_moves` functions, along with printing the initial board, as per the instructions.



````python
import numpy as np

# 1. Define a 3x3 game board using a list of lists. Initialize all cells to represent an empty state.
# Using ' ' for empty cells
board = [[' ' for _ in range(3)] for _ in range(3)]

# 2. Implement check_win function
def check_win(board, player):
    # Check rows
    for row in board:
        if all([cell == player for cell in row]):
            return True

    # Check columns
    for col in range(3):
        if all([board[row][col] == player for row in range(3)]):
            return True

    # Check main diagonal
    if all([board[i][i] == player for i in range(3)]):
        return True

    # Check anti-diagonal
    if all([board[i][2 - i] == player for i in range(3)]):
        return True

    return False

# 3. Implement check_draw function
def check_draw(board):
    # A draw occurs if no player has won and there are no available moves
    if not get_available_moves(board) and not check_win(board, 'X') and not check_win(board, 'O'):
        return True
    return False

# 4. Implement get_available_moves function
def get_available_moves(board):
    moves = []
    for r in range(3):
        for c in range(3):
            if board[r][c] == ' ':
                moves.append((r, c))
    return moves

# 5. Print an initial empty game board
print("Initial Empty Game Board:")
for row in board:
    print("|" + "|".join(row) + "|")
````

## Implement Minimax Function

### Subtask:
Write a Python function for the Minimax algorithm that takes the current game state, depth, and a boolean indicating if it's the maximizing player's turn. This function should recursively explore game states and return the optimal score.


**Reasoning**:
I will implement the `minimax` function as described, incorporating the base cases for win/loss/draw, and the recursive logic for both maximizing and minimizing players, ensuring to use the previously defined helper functions for game state checks.



````python
def minimax(board, depth, is_maximizing_player):
    # Base cases for win/loss/draw
    if check_win(board, 'X'):  # Maximizing player wins
        return 10 - depth
    if check_win(board, 'O'):  # Minimizing player wins
        return depth - 10
    if check_draw(board):  # Draw
        return 0

    if is_maximizing_player:
        max_eval = -float('inf')
        for r, c in get_available_moves(board):
            # Make the move
            board[r][c] = 'X'
            # Recursively call minimax
            eval = minimax(board, depth + 1, False)
            # Undo the move
            board[r][c] = ' '
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        for r, c in get_available_moves(board):
            # Make the move
            board[r][c] = 'O'
            # Recursively call minimax
            eval = minimax(board, depth + 1, True)
            # Undo the move
            board[r][c] = ' '
            min_eval = min(min_eval, eval)
        return min_eval

print("Minimax function defined successfully.")
````
